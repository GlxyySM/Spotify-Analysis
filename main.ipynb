```python
import pandas as pd
import plotly.express as pl

tabela = pd.read_csv('top50.csv', encoding='latin1')

display(tabela)
```

```python
tabela = tabela.rename(columns={'Beats.Per.Minute': 'BPM',
                               'Loudness..dB..':'Loudness',
                               'Track.Name': 'Song Name',
                               'Artist.Name': 'Artist Name',
                               'Valence.': 'Valence'})

tabela = tabela[['Song Name', 'Artist Name', 'BPM', 'Energy', 'Danceability', 'Loudness', 'Liveness', 'Valence']]
display(tabela)
```

```python
tabela = tabela[['BPM', 'Energy', 'Danceability', 'Loudness', 'Valence']]
tabela['BPM'] = tabela['BPM'] / 2
tabela['Loudness'] = (tabela['Loudness'] + 11) * 10
display(tabela.describe())
```

```python
tab_med = tabela.mean()
graph = pl.line_polar(r=tab_med.values, theta=tab_med.index, range_r=[0, 100], line_close=True)

display(graph)
```
